```python
new_df=df.drop(['id','host_name','host_id','name','number_of_reviews','last_review','reviews_per_month','calculated_host_listings_count','availability_365','latitude','longitude'],axis=1)
new_df.head()
```

```python
p=(new_df['price'] / new_df['minimum_nights']).rename('Price_per_nights')
```

```python
new_df=pd.concat([new_df,p],axis=1)
```

```python
new_df
```

```python
min,max=new_df.Price_per_nights.quantile([0.01,0.99])
min,max
```

```python
no_outlier=new_df[(new_df.Price_per_nights>min) & (new_df.Price_per_nights<max)]
no_outlier
```
